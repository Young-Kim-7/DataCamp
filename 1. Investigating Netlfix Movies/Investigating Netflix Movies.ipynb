```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
```

```python
# 1. Load and inspect the Netflix data
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
# 2. Subset the DataFrame for "Movies"
netflix_subset = netflix_df[netflix_df["type"]=="Movie"]
```

```python
# 3. Subset the columns of the new DataFrame
netflix_movies= netflix_subset[["title","country","genre","release_year","duration"]]
netflix_movies.head()
```

```python
# 4. Filter the DataFrame by Movie duration
short_movies = netflix_movies[netflix_movies["duration"]<60]
short_movies.head(20)
```

```python
# 5. Assign colors to movie genres
colors = []
for ind, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("red")
    elif row["genre"] == "Documentaries":
        colors.append("green")
    elif row["genre"] == "Stand-up":
        colors.append("blue")
    else:
        colors.append("grey")
# colors[:10] (For checking out)
```

```python
# 6. Plot the movie duration by year using the genre colors created
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies["release_year"],netflix_movies["duration"], color=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
